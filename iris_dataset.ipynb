```python
from sklearn import datasets

iris = datasets.load_iris()
```

```python
type(iris)
```

```python
print(iris)
```

```python
import matplotlib.pyplot as plt

_, ax = plt.subplots()
scatter = ax.scatter(iris.data[:, 2], iris.data[:, 3], c=iris.target)
ax.set(xlabel=iris.feature_names[2], ylabel=iris.feature_names[3])
_ = ax.legend(
    scatter.legend_elements()[0], iris.target_names, loc="lower right", title="Classes"
)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1,4, figsize=(20,4))
for i in range(4):
    scatter = ax[i].scatter(iris.data[:, 0], iris.data[:, i], c=iris.target)
    ax[i].set(xlabel=iris.feature_names[0], ylabel=iris.feature_names[i])
    _ = ax[i].legend(
        scatter.legend_elements()[0], iris.target_names, loc="lower right", title="Classes"
    )
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1,4, figsize=(20,4))
for i in range(4):
    scatter = ax[i].scatter(iris.data[:, 1], iris.data[:, i], c=iris.target)
    ax[i].set(xlabel=iris.feature_names[1], ylabel=iris.feature_names[i])
    _ = ax[i].legend(
        scatter.legend_elements()[0], iris.target_names, loc="lower right", title="Classes"
    )
```

```python
# unused but required import for doing 3d projections with matplotlib < 3.2
import mpl_toolkits.mplot3d  # noqa: F401

from sklearn.decomposition import PCA

fig = plt.figure(1, figsize=(8, 6))
#ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)
ax = fig.add_subplot(111,)

X_reduced = PCA(n_components=2).fit_transform(iris.data)
ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
#    X_reduced[:, 2],
    c=iris.target,
    s=40,
)

ax.set_title("First three PCA dimensions")
ax.set_xlabel("1st Eigenvector")
#ax.xaxis.set_ticklabels([])
ax.set_ylabel("2nd Eigenvector")
#ax.yaxis.set_ticklabels([])
#ax.set_zlabel("3rd Eigenvector")
#ax.zaxis.set_ticklabels([])

plt.show()
```
